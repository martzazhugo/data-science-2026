```python
# =================================================================
# PERSIAPAN & IMPORT DATA
# =================================================================
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("--- Dataset Berhasil Dimuat ---")
display(df.head())

# =================================================================
# Q1: JUMLAH BARIS, KOLOM, DAN NAMA KOLOM
# =================================================================
print("\n[Q1] Informasi Struktur Data:")
rows, cols = df.shape
print(f"Jumlah Baris: {rows}")
print(f"Jumlah Kolom: {cols}")
print(f"Daftar Kolom : {df.columns.tolist()}")

# =================================================================
# Q2: MISSING VALUES (JUMLAH & PERSENTASE)
# =================================================================
print("\n[Q2] Analisis Missing Values:")
missing_data = pd.DataFrame({
    'Jumlah Missing': df.isnull().sum(),
    'Persentase (%)': (df.isnull().sum() / len(df)) * 100
})
# Menampilkan hanya yang memiliki missing value
display(missing_data[missing_data['Jumlah Missing'] > 0])

# =================================================================
# Q3: PERSENTASE PENUMPANG SELAMAT
# =================================================================
# Menggunakan kolom Survived (1 = Selamat, 0 = Meninggal)
survival_rate = df['Survived'].mean() * 100
print(f"\n[Q3] Persentase Penumpang Selamat: {survival_rate:.2f}%")

# =================================================================
# Q4: FILTER WANITA KELAS 1 (PCLASS=1)
# =================================================================
female_c1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
jml_female_c1 = len(female_c1)
avg_age_female_c1 = female_c1['Age'].mean()

print(f"\n[Q4] Analisis Wanita Kelas 1:")
print(f"- Jumlah Penumpang: {jml_female_c1}")
print(f"- Rata-rata Usia  : {avg_age_female_c1:.2f} tahun")

# =================================================================
# Q5: TINGKAT KESELAMATAN PER KELAS
# =================================================================
print("\n[Q5] Tingkat Keselamatan Berdasarkan Kelas:")
class_survival = df.groupby('Pclass')['Survived'].mean() * 100
print(class_survival)
print(f"Kelas dengan tingkat keselamatan tertinggi: Kelas {class_survival.idxmax()}")

# =================================================================
# Q6: ANALISIS TAMBAHAN (KESELAMATAN BERDASARKAN JENIS KELAMIN)
# =================================================================
print("\n[Q6] Analisis Tambahan (Gender vs Survival):")
gender_survival = df.groupby('Sex')['Survived'].mean() * 100
print(gender_survival)

# Interpretasi Singkat:
# Terlihat perbedaan drastis di mana tingkat keselamatan wanita jauh lebih tinggi (74%)
# dibandingkan pria (18%). Hal ini mencerminkan prioritas evakuasi "wanita dan anak-anak".
```
